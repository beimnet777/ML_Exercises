```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target']=data.target
```

```python
df.head(51)
```

```python
from matplotlib import pyplot as plt
```

```python
df1=df.loc[df.target==0]
df2=df.loc[df.target==1]
df3=df.loc[df.target==2]
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color="red")
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color="blue")
plt.scatter(df3['sepal length (cm)'],df3['sepal width (cm)'],color="green")
```

```python
model=KNeighborsClassifier(n_neighbors=3)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(['target'],axis='columns'),df.target)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,model.predict(X_test))
```

```python
import joblib
joblib.dump(model,'KNN.joblib')
```
