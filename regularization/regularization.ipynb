```python
import pandas as pd
df=pd.read_csv('house.csv')
df.isna().sum()
```

```python
df.head(10)
```

```python
df['Lattitude'].fillna(df['Lattitude'].mean(),inplace=True)
df['Longtitude'].fillna(df['Longtitude'].mean(),inplace=True)
df['Price'].fillna(df['Price'].mean(),inplace=True)
df['BuildingArea'].fillna(df['BuildingArea'].mean(),inplace=True)
df['Bedroom2'].fillna(df['Bedroom2'].mean(),inplace=True)
df['Landsize'].fillna(df['Landsize'].mean(),inplace=True)
df['Bathroom'].fillna(df['Bathroom'].mean(),inplace=True)
df['Car'].fillna(df['Car'].mean(),inplace=True)
df['YearBuilt'].fillna(df['YearBuilt'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
len(df)
```

```python
df['BuildingArea']
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df.loc[df['CouncilArea']!='Yarra City Council']
```

```python
df.drop('Postcode',axis=1,inplace=True)
df.drop(['Address','SellerG'],axis=1,inplace=True)
df.drop('Regionname',axis='columns',inplace=True)
df
```

```python
df.head()
```

```python
dummies=pd.get_dummies(df.Type)
dummies.drop('u',axis=1,inplace=True)
dummies
df.drop('Type',axis=1,inplace=True)
df=pd.concat([df,dummies],axis=1)
df
```

```python


dummies=pd.get_dummies(df.Method)
dummies
df.drop('Method',axis='columns',inplace=True)
df
```

```python
df.drop(['Date','Suburb','CouncilArea'],axis=1,inplace=True)
df
```

```python
from sklearn.linear_model import Lasso,Ridge
model1=Lasso(alpha=7,max_iter=1000,tol=0.01)
model2=Ridge(alpha=7,max_iter=1000,tol=0.01)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('Price',axis=1),df.Price,test_size=0.2)
```

```python
X_train
```

```python

model2.fit(X_train,y_train)
```

```python
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

```python
model2.score(X_test,y_test)
```
